```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
filename = 'white_dwarf_data.csv'
data = pd.read_csv(filename)

# Constants
G = 6.67430e-8  
solar_mass = 1.98847e30 #kg
earth_radius = 6.3781e6 #m

def convert_log_g_to_radius(mass, log_g):
    
    mass_g = mass * solar_mass 
    g = 10**log_g  # Convert log(g) to g
    
    radius_cm = np.sqrt(G * mass_g / g)
    radius = radius_cm / earth_radius
    
    return radius

radii = data.apply(lambda row: convert_log_g_to_radius(row['mass'], row['logg']), axis=1).to_numpy()
masses = data.loc[:,"mass"].to_numpy()

plt.figure()
plt.scatter(radii, masses)
plt.xlabel('Radius')
plt.ylabel('Mass')
plt.title('Radius vs Mass')
plt.show()
```
